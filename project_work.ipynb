# Predicting  AGE and PAY : Linear Regression
## 5.1 80-20 Split of Training and Hold-Out Data ¶


```python
from sklearn.model_selection import train_test_split


train, test = train_test_split(df, test_size=0.2, random_state=4) 
```

```python
from IPython.display import Markdown, display
def printmd(string):
    display(Markdown(string))
```

```python
results1 = smf.ols('AGE ~ PAY_0 + PAY_2 + PAY_3 + PAY_4 ', data=df).fit()
print(results1.summary())
```

```python
y = train["AGE"]
cols = ["PAY_0","PAY_2","PAY_3","PAY_4"]

X=train[cols]
```

```python
regr = linear_model.LinearRegression()
regr.fit(X,y)

rscore = regr.score(X, y)
print("rscore=",rscore)
r_coef = regr.coef_
printmd("**Estimated coefficients for the linear regression problem. If multiple targets are passed during the fit (y 2D), this is a 2D array of shape (n_targets, n_features), while if only one target is passed, this is a 1D array of length n_features**")
print("r_coef=",r_coef)
rintercept = regr.intercept_
print("rintercept=",rintercept)
```

```python
ytrain_pred = regr.predict(X)
print("In-sample Mean squared error: %.2f"
      % mean_squared_error(y, ytrain_pred))
```

```python
ytest = test["AGE"]
cols = ["PAY_0","PAY_2","PAY_3","PAY_4"]

Xtest=test[cols]
```

```python
ypred = regr.predict(Xtest)
print("Out-of-sample Mean squared error: %.2f"
      % mean_squared_error(ytest, ypred))
```

```python
from sklearn.metrics import cohen_kappa_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

print("ytest:" , len(ytest))
print("ypred :" , len(ypred))

# target_names = ['class 0', 'class 1', 'class 2']
# print(classification_report(ytest, ytrain_pred, target_names=target_names))
# confusion_matrix(ytest, ypred)
# cohen = cohen_kappa_score(ytest, ypred)
# print("cohen_kappa_score: " + cohen)
```

```python

printmd("**The Lasso is a linear model that estimates sparse coefficients. It is useful in some contexts due to its tendency to prefer solutions with fewer non-zero coefficients, effectively reducing the number of features upon which the given solution is dependent. For this reason Lasso and its variants are fundamental to the field of compressed sensing. Under certain conditions, it can recover the exact set of non-zero coefficients **")
from sklearn import linear_model
clf = linear_model.Lasso(alpha=0.1)
clf.fit(X,y)
clf_score = clf.score(X, y)
print("clf_score=",clf_score)
clf_coef = clf.coef_
print("clf_coef=",clf_coef)
clf_intercept = clf.intercept_
print("clf_intercept=",clf_intercept)

lasso_ytrain_pred = clf.predict(X)
print("LASSO In-sample Mean squared error: %.2f"
      % mean_squared_error(y, lasso_ytrain_pred))

lasso_ypred = clf.predict(Xtest)
print("LASSO Out-of-sample Mean squared error: %.2f"
      % mean_squared_error(ytest, lasso_ypred))
```
